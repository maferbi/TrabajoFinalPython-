```python
#Se importan los paquetes de Python
import pandas as pd
import numpy as np 
import altair as alt

```

```python
#La información del contenido en Netflix más visualizado en EEUU entre 2020 y marzo de 2022 proviene de esta dirección web en Kaggle-> https://www.kaggle.com/prasertk/netflix-daily-top-10-in-us
#df tiene 10 columnas y 7100 filas
df = pd.read_csv (r'C:\Users\ferna\Documents\2022- Verano\Python\Netflix.csv')
print (df)
```

```python
#Se realiza esta acción para asegurar que los valores de la columna As of estén en formato de fecha 
df['As of'] = pd.to_datetime(df['As of'])
```

```python
#Se configura la fecha de inicio y final con el fin de acotar la información de los usuarios.
##Originalmente se planeaba utilizar los datos de visualización durante el primer confinamiento obligatorio durante la pandemia
##sin embargo EEUU no tuvo un periodo de confinamiento a nivel nacional. Por lo que, se optó por el uso de los datos desde 
##el inicio del "lockdown" en algunas de las ciudades más grandes de ese país hasta exactamente un año después.  
start_date = '13-03-2020'
end_date = '13-03-2021'
```

```python
#Se utiliza la función más para seleccionar la visualización de streaming
mask = (df['As of'] > start_date) & (df['As of'] <= end_date)
```

```python
#Se modifica el df original 
df = df.loc[mask]
df
```

```python
#Se renombran las columnas en español
print(df.rename(columns={'As of': 'Desde', 'Rank': 'Posición', 'Year to Date Rank': 'Posición del año hasta la fecha', 'Last Week Rank': 'Posición de la semana anterior', 'Title': 'Título', 'Type': 'Tipo','Netflix Exclusive': 'Exclusivo de Netflix', 'Netflix Release Date': 'Fecha de lanzamiento en Netflix', 'Days In Top 10': 'Días en el top 10', 'Viewership Score': 'Puntuación de audiencia'}, inplace=True))
```

```python
#El df se visualiza de esta manera tras el renombramiento de las columnas
df
```

```python
#Se reemplaza en la columna tipo, el tipo de contenido en español
df["Tipo"].replace({"TV Show": "Show de TV", "Movie": "Película","Concert/Performance": "Concierto",  "Stand-Up Comedy": "Comedia Stand-Up"}, inplace=True)
print(df)
```

```python
#El df se visualiza de esta manera tras el reemplazo en la columna Tipo
df
```

```python
#Se reemplaza en la columna Exclusivo de Netflix el valor yes en español
df["Exclusivo de Netflix"].replace({"Yes": "Si"}, inplace=True)
print(df)
```

```python
#El df se visualiza de esta manera tras el reemplazo en la columna Exclusivo de Netflix
df
```

```python
#Se decide utilizar solamente la información correspondiente al Contenido Original de Netflix que se visualizará tras el uso de dropna
df.dropna(inplace = True)
```

```python
#El df se visualiza de esta manera tras eliminar los valores missing de Exclusivo de Netflix
df
```

```python
#Se configura la tabla con la función mark_point debido a que tras evaluar las otras posibilidades, 
#esta es entendida como la mejor para la visualización del consumo de contenido a lo largo del tiempo.
#En el eje X, se ubican los títulos de Contenido Original de Netflix: las series, películas, especiales de comedia de Stand-Up y especiales de Conciertos.
#Por otro lado, en el eje Y se ubica las fechas de visualización del contenido en formato de días y meses.
#Se utilizó la puntuación de audiencia respecto al tamaño de los puntos para poder determinar la popularidad de una manera más rápida.
#Además, a través de tooltip se puede conocer la posición, posición del año hasta la fecha y los días en el top 10 del contenido. 
#Finalmente se adicionó la interactividad con la tabla y la configuración del título de la misma.


chart= alt.Chart(df).mark_point().encode(
    alt.X("Título", title= "Título de Contenido Original" ),
    alt.Y("Desde", title= "Fecha de visualización"), color= "Tipo",
    size= alt.Size ("Puntuación de audiencia"),
    tooltip = ["Posición","Posición del año hasta la fecha", "Días en el top 10"]
).interactive().properties(
    title={"text":["Consumo de Contenido Original de Netflix en Estados Unidos"], "subtitle":["entre abril de 2020 y marzo de 2021",
                   "Fuente: Kaggle"],
           
      "color": "Black",
      "subtitleColor": "Grey"
    }
)

chart.configure_title(
    fontSize=20,
    font='Courier',
    anchor='start',
    subtitleFontSize=15,
    subtitleFont='Courier',
    
)
```
